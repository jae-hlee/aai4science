```python
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt

timestamps = [
    "11/6/2025 14:16:14",
    "11/6/2025 14:25:56",
    "11/6/2025 14:30:41",
    "11/6/2025 14:32:41",
    "11/6/2025 14:35:32",
    "11/6/2025 14:42:08",
    "11/6/2025 14:46:04",
    "11/6/2025 14:47:13",
    "11/6/2025 14:48:04",
    "11/6/2025 14:48:49",
    "11/6/2025 14:49:20",
    "11/6/2025 14:50:41",
    "11/6/2025 14:51:45",
    "11/6/2025 14:52:03",
    "11/6/2025 14:53:07",
    "11/6/2025 14:53:09",
    "11/6/2025 14:53:23",
    "11/6/2025 14:53:49",
    "11/6/2025 14:54:05",
    "11/6/2025 14:55:14",
    "11/6/2025 14:55:30",
    "11/6/2025 14:55:40",
    "11/6/2025 14:56:59",
    "11/6/2025 14:57:15",
    "11/6/2025 14:57:59",
    "11/6/2025 14:58:04",
    "11/6/2025 14:58:20",
    "11/6/2025 14:58:28",
    "11/6/2025 14:58:40",
    "11/6/2025 14:58:55",
    "11/6/2025 14:58:59",
    "11/6/2025 15:00:17",
    "11/6/2025 15:00:24",
    "11/6/2025 15:00:27",
    "11/6/2025 15:00:31",
    "11/6/2025 15:00:35",
    "11/6/2025 15:00:41",
    "11/6/2025 15:00:59",
    "11/6/2025 15:04:18",
    "11/6/2025 15:04:28",
    "11/6/2025 15:09:12",
    "11/6/2025 15:18:07",
    "11/6/2025 15:22:37",
    "11/6/2025 15:27:21",
]

finish_times = [datetime.strptime(t, "%m/%d/%Y %H:%M:%S") for t in timestamps]
start_time = datetime.strptime("11/6/2025 14:00:00", "%m/%d/%Y %H:%M:%S")

minutes_to_finish = [
    (t - start_time).total_seconds() / 60 for t in finish_times
]

df = pd.DataFrame({
    "Participant": range(1, len(minutes_to_finish) + 1),
    "Minutes to Finish": minutes_to_finish
})

participant_percent = (
    df["Participant"] - 1
) / (len(df) - 1) * 100

plt.figure(figsize=(10, 5))
plt.plot(
    participant_percent,
    df["Minutes to Finish"],
    marker='o'
)

plt.xlabel("Participants (%)", fontsize=16)
plt.ylabel("Time to Finish (minutes)", fontsize=16)

plt.xticks(fontsize=16)
plt.yticks(fontsize=16)

plt.grid(True)
plt.tight_layout()
plt.show()
```
