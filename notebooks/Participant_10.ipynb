TASK 1: **The Wonderful Weather in Baltimore **

```python
import requests

# API configuration
api_key = "sk-"
location = "Baltimore"
url = f"https://atomgpt.org/weather?location={location}&APIKEY={api_key}"

# Make the API call
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    weather_data = response.json()
    print(f"Weather in {location}:")
    print(weather_data)
else:
    print(f"Error: {response.status_code}")
    print(response.json())
```

Task 3: MgB2 Updates from ArXiv (10 latest

```python

import requests
import json

# API configuration
api_key = "sk-"
query = "MgB2"
url = f"https://atomgpt.org/arxiv?query={query}&APIKEY={api_key}"

# Make the API call
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    papers = response.json()
    print(f"{'='*80}")
    print(f"Latest 10 Papers on {query} from arXiv")
    print(f"{'='*80}\n")

    # Display the first 10 papers in a nice format
    if isinstance(papers, list):
        for i, paper in enumerate(papers[:10], 1):
            print(f"Paper {i}:")
            print(f"{'-'*80}")
            if isinstance(paper, dict):
                print(f"Title: {paper.get('title', 'N/A')}")
                print(f"Authors: {paper.get('authors', 'N/A')}")
                print(f"Published: {paper.get('published', 'N/A')}")
                print(f"Summary: {paper.get('summary', 'N/A')[:200]}...")
                print(f"Link: {paper.get('link', 'N/A')}")
            else:
                print(paper)
            print(f"\n")
    else:
        print(json.dumps(papers, indent=2))

    print(f"{'='*80}")
else:
    print(f"Error: {response.status_code}")
    print(response.json())
```
